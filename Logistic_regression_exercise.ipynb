```python
import pandas as pd #import pandas
titanic_df=pd.read_excel('/Users/Carol/Desktop/titanic_data.xlsx') #read data from excel
titanic_df #show the data
```

```python
age_null=pd.isnull(titanic_df["Age"]) #pick up the lines where the Age is NaN
titanic_df[age_null].shape[0]
```

```python
titanic_df=titanic_df.dropna(subset=['Age']) #drop all the passengers from the dataset who miss the age information
titanic_df=titanic_df.reset_index(drop=True) #reorder the index of each row
titanic_df
```

```python
X=titanic_df.drop(columns=['Name','Sex','Ticket','Cabin','Embarked','Survived','PassengerId'])
y=titanic_df['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
logreg=LogisticRegression()
solvers = ['liblinear','lbfgs']
penalty = ['l1', 'l2']
c_values = [100, 10, 1.0, 0.1, 0.01]
param_grid = dict(solver=solvers,penalty=penalty,C=c_values)
cv=StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
grid=GridSearchCV(estimator=logreg, param_grid=param_grid, cv=cv, n_jobs=-1, verbose=2)
grid.fit(X,y)
print("The best parameters are %s with a score of %0.2f" %(grid.best_params_,grid.best_score_))
```

```python
#C means the inverse of the regularization strength. The stronger the regularization (and thus the smaller C is), the smaller the coefficients in the model will be and for lasso the fewer the coefficients it will have. Shrinking the coefficients too much or eliminating too many variables can weaken the model so how much regularization is needed is determined here using cross-validation.
```

```python
#The penalty refers to regularization penalties that shrink or eliminate variable-coefficients in the model. The l1 penalty refers to lasso regularization which causes some of the feature-variable coefficients to go to 0 if they donâ€™t contribute as much to the model as the other, more significant variables. The l2 penalty refers to ridge regression which shrinks the coefficients but never pushes them all the way to 0. Since lasso regularization simplifies the model it might seem that it would be preferable, but it will only outperform ridge regression if there are differences in how much the variables contribute. If all the variables contribute equally to the model, then ridge regression will outperform the lasso
```
